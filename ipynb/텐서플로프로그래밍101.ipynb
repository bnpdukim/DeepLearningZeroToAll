# Machine learning

```python
import tensorflow as tf
hello = tf.constant('Hello, TensorFlow!')
hello2 = tf.constant('Hello, TensorFlow!')
print(hello)
print(hello2)
```

```python
a = tf.constant(10)
b = tf.constant(32)
c = tf.add(a,b)
print(c)
```

```python
sess = tf.Session()
print(sess.run(hello))
print(sess.run(c))
print(sess.run([a,b,c]))
sess.close()
```

```python
X = tf.placeholder(tf.float32, [None, 3], name="X") #[[1,2,3],[1,2,3],[1,2,3],[1,2,3]]
print(X)
```

```python
W = tf.Variable(tf.random_normal([3,2]))  # random_normal : 정규분포의 무작위 값으로 초기화
b = tf.Variable(tf.random_normal([2,1]))
expr = tf.matmul(X,W) + b
```

```python
x_data = [[1,2,3], [4,5,6]] # [2,3]
sess = tf.Session()
sess.run(tf.global_variables_initializer())

print('===x===')
print(x_data)
print('===W===')
print(sess.run(W))
print('===b===')
print(sess.run(b))
print('===expr===')
print(sess.run(expr,feed_dict={X: x_data}))
sess.close()
```

# linear regression

```python
x_data = [1,2,3]
y_data = [1,2,3]
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0)) # -1.0~1.0 균등분포
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")

hypothesis = W * X + b # W : 가중치, b : 편향
cost = tf.reduce_mean(tf.square(hypothesis - Y)) # http://dlwodus.tistory.com/27
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.1) # 경사 하강법
train_op = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(100):
        _, cost_val = sess.run([train_op, cost], feed_dict={X: x_data, Y: y_data})
        print(step, cost_val, sess.run(W), sess.run(b))
    print("X: 5, Y:", sess.run(hypothesis, feed_dict={X:5}))
    print("X: 2.5, Y:", sess.run(hypothesis, feed_dict={X:2.5}))
```
